#### Unsupervised learning (PCA or KMeans)

Ledningen vill inte bara ha en modell – de vill också förstå om det finns naturliga grupper av områden i datan. Om sådana grupper finns kan det hjälpa till att:
segmentera områden (t.ex. “typområden” som liknar varandra),
upptäcka ovanliga områden (avvikare),
och få en enklare överblick över datan innan man tar beslut.

Uppdrag: 
Du ska därför undersöka om datan verkar innehålla struktur genom att använda PCA eller KMeans på ett rimligt urval av X-variabler (inte target).

Syfte - Att se om vi kan:
- sammanfatta datan i ett enklare “mönster” (PCA), eller
- hitta grupper av liknande områden (KMeans),
- och diskutera hur detta skulle kunna användas som beslutsstöd.

Krav:

Använd endast X-variabler (inte target), och motivera kort vilka features du inkluderade. Kom ihåg att metoderna är känsliga för skalning.

```python
X_cluster = add_ratio_features(X_train)

imputer = SimpleImputer(strategy="median")          # hanterar NaN
scaler = StandardScaler()                       

X = add_ratio_features(X_train)

X_imputed = imputer.fit_transform(X[num_features])
X_scaled = scaler.fit_transform(X_imputed)

kmeans = KMeans(n_clusters=6, random_state=42)

clusters = kmeans.fit_predict(X_scaled)

# assign clusters to the dataframe
X_train["clusters"] = clusters
```

### Plotting geographical patterns

To see if there are any geographical patterns This gives a first glance at spatial patterns.

```python
plt.scatter(
    X_cluster["longitude"],
    X_cluster["latitude"],
    c=X_cluster["cluster"],
    cmap="viridis",
    s=10,
    alpha=0.6
)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("KMeans Clusters of California Housing")

plt.show()
```

### Comparing cluster statistics (mean values)

The best way to interpret clusters is to check mean values of the features per cluster. That gives a sense of what differentiates each cluster:

Look at patterns like:

- Which clusters have high median_income?

- Which clusters have high population_per_household?

- Are certain clusters concentrated in specific lat/lon regions?

```python
cluster_summary = X_cluster.groupby("clusters")[num_features].mean()
print(cluster_summary)
```
